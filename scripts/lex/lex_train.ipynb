# Dataset

```python
df = pd.read_parquet(DATA_PATH.joinpath('dia-final.parquet'))
```

## Train Test Split

```python
sampled = df['chunk_id'].drop_duplicates().sample(frac=0.8).to_numpy()
train = df[df['chunk_id'].isin(sampled)]
test = df[~df['chunk_id'].isin(sampled)]
```

```python
for df in [train, test]:
    ids_oton = {v:k for k,v in enumerate(df['chunk_id'].unique())}
    df['chunk_id'] = df['chunk_id'].apply(lambda x: ids_oton[x])
```
